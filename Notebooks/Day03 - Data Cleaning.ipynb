### All arrays are Unique

```python
df["lunch"] = df["lunch"].str.replace("standard", "high")
```

```python
df["lunch"] = df["lunch"].str.replace("free/reduced", "poor")
```

```python
df["race/ethnicity"] = df["race/ethnicity"].str.replace("group A", "Asians")
```

```python
df["race/ethnicity"] = df["race/ethnicity"].str.replace("group B", "Americans")
```

```python
df["race/ethnicity"] = df["race/ethnicity"].str.replace("group C", "Europeans")
```

```python
df["race/ethnicity"] = df["race/ethnicity"].str.replace("group D", "Africans")
```

```python
df["race/ethnicity"] = df["race/ethnicity"].str.replace("group E", "Arabians")
```

```python
df = df.rename(columns={"lunch" : "lunch quality"})
```

```python
df = df.rename(columns={"race/ethnicity" : "race"})
```

### Items in the Lunch column were renamed from:
- "Standard" to "High"
- "free/reduced" to "Poor"
### Meanwhile items in the Race/Ethnicity column were renamed from:
- Group A to Asians
- Group B to Americans
- Group C to Europeans
- Group D to Africans
- Group E to Arabians
### Column names were changed from:
- Lunch to Lunch Quality
- Race/Ethnicity to Race
