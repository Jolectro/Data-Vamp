```python
sns.boxplot(x='gender', y='math score', data=df)
plt.title('Which Gender performs well in Math')
plt.show()
```

```python
sns.boxplot(data=df, x='gender', y='reading score')
plt.title('Which Gender loves reading?')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='parental level of education')
plt.title('Parents Educational Status')
plt.show()
```

```python
sns.countplot(data=df, x='lunch')
plt.title('How many ate well?')
plt.show()
```

```python
sns.countplot(data=df, x='test preparation course')
plt.title('How many prepared for the tests?')
plt.show()
```

```python
whelmed = df.select_dtypes(include='number')
plt.figure(figsize=(12,8))
sns.heatmap(whelmed.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
